```python
import configparser
import jqdatasdk as jq

config = configparser.ConfigParser()
config.read("/home/jovyan/configs/dfdata.ini")
jq.auth(config['jqdata']['id'], config['jqdata']['password'])
```

```python
# 查看当天剩余行数
jq.get_query_count()
```

## 标的信息

```python
# 所有期货合约
futures_contracts = jq.get_all_securities(['futures'])
futures_contracts.tail(5)
```

## 行情数据

```python
df = jq.get_price(
    'A0501.XDCE', 
    start_date='2004-12-01',
    end_date='2006-1-22',
    frequency='daily', 
    fields=['open', 'close', 'low', 'high', 'volume', 'money', 'factor', 'high_limit','low_limit', 'avg', 'pre_close', 'paused', 'open_interest'])
df
```

```python
df2 = jq.get_price(['ZN2104.XSGE', 'ZN9999.XSGE'], start_date='2020-05-10', end_date='2020-05-10', frequency='daily', fields=None, skip_paused=False, fq='pre', count=None, panel=True, fill_paused=True)
df2
```
